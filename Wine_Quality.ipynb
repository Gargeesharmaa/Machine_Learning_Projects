```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
dataset = pd.read_csv('/content/WineQT.csv')
dataset.head()
```

```python
dataset.shape
```

```python
dataset['quality'].value_counts()
```

```python
dataset['quality'] = dataset['quality'].apply(lambda x: 1 if x>5 else 0)
```

```python
dataset['quality'].value_counts()
```

```python
dataset.isnull().sum()
```

```python
x= dataset.drop('quality', axis=1)
y = dataset['quality']
x_train, x_test, y_train, y_test = train_test_split(x,y, random_state=8, stratify=y, test_size=0.3 )
```

```python
classifier = RandomForestClassifier(n_estimators=110)
classifier.fit(x_train,y_train)
```

```python
classifier.predict(x_test)
```

```python
classifier.score(x_test, y_test)
```
