```python
import pandas as pd
import numpy as np
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
```

```python
#DATACOLLECTION
diabetic_dataset = pd.read_csv('diabetes.csv')
diabetic_dataset.head()
diabetic_dataset.shape
```

```python
diabetic_dataset.value_counts('Outcome')
```

```python
#DATA PREPROCESSING
diabetic_dataset.groupby('Outcome').mean()
```

```python

X = diabetic_dataset.drop(columns = 'Outcome', axis = 1)
Y = diabetic_dataset['Outcome']
print(X)
print(Y)
```

```python
#DATA STANDARDIZATION
scaler = StandardScaler()
standardize = scaler.fit_transform(X)
print(standardize)
```

```python
X = standardize
print(X)
```

```python
#DATA SEPARATE IN TRAIN AND TEST DATA
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, stratify = Y, random_state = 2)
print(X_train.shape, X_test.shape, Y_train.shape, Y_test.shape)
```

```python
classifier = svm.SVC(kernel='linear')
classifier.fit(X_train, Y_train)
```

```python
x_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(x_train_prediction, Y_train)
print(training_data_accuracy)
```
