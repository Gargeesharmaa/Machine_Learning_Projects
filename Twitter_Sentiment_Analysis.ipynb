```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import BernoulliNB
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
df = pd.read_csv('/content/training.1600000.processed.noemoticon.csv.zip', encoding= 'latin-1', header=None)
df = df[[0,5]]
df.columns = ['polarity', 'text']
df.head()
```

```python
df['polarity']= df['polarity'].map({0:0, 4:1})
df.head()
```

```python
df.isnull().sum()
```

```python
def clean_text(text):
  text = text.lower()
  return text
df['text']= df['text'].apply(clean_text)
df.head()
```

```python
x = df['text']
y= df['polarity']
x_train, x_test, y_train, y_test = train_test_split(x,y, random_state= 42, test_size= 0.2)
```

```python
vectorizer = TfidfVectorizer(max_features = 5000, ngram_range= (1,2))
x_train_vectorizer = vectorizer.fit_transform(x_train)
x_test_vectorizer = vectorizer.fit_transform(x_test)
```

```python
bnb = BernoulliNB()
bnb.fit(x_train_vectorizer, y_train)
bnb_pred = bnb.predict(x_test_vectorizer)
print("bernoulliNB accuracy score",accuracy_score(y_test, bnb_pred))
print("bernoulliNB classification report", classification_report(y_test, bnb_pred))
print("bernoulliNB conlusion score ", confusion_matrix(y_test, bnb_pred))
```

```python
svm = LinearSVC(max_iter=1000)
svm.fit(x_train_vectorizer, y_train)
svm_pred = svm.predict(x_test_vectorizer)
print("Svm accuracy score",accuracy_score(y_test, svm_pred))
print("Svm classification report", classification_report(y_test, svm_pred))
print("Svm conlusion score ", confusion_matrix(y_test, svm_pred))
```

```python
Logrex = LogisticRegression(max_iter= 100)
Logrex.fit(x_train_vectorizer, y_train)
logrex_pred = Logrex.predict(x_test_vectorizer)
print("LogisticRegression accuracy score",accuracy_score(y_test, logrex_pred))
print("LogisticRegression classification report", classification_report(y_test, logrex_pred))
print("LogisticsRegresiion conlusion score ", confusion_matrix(y_test, logrex_pred))
```
